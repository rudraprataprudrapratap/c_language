FOREST COVER TYPE PREDICTOR

```python
# -----------------------------------------------
# 1. Import All Necessary Libraries
# -----------------------------------------------
print("Step 1: Importing libraries...")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_covtype
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC  # Using LinearSVC for speed.
                                 # Standard SVC(kernel='rbf') is too slow for this data size.
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

print("Libraries imported successfully.\n")
```

```python
# -----------------------------------------------
# 2. Load All The Data Set
# -----------------------------------------------
print("Step 2: Loading dataset...")
# Fetch the dataset
covtype_data = fetch_covtype()
```

```python
# Convert to a pandas DataFrame for easier preprocessing
df = pd.DataFrame(data=covtype_data.data, columns=covtype_data.feature_names)
df['target'] = covtype_data.target

print(f"Full dataset shape: {df.shape}")
```

```python
# --- CRITICAL STEP: Subsampling ---
# The dataset is too large for a standard SVM. We will take a stratified sample.
print("Subsampling data to 30,000 rows to make SVM training feasible...")

# Calculate the desired sample size and ratio for stratified sampling
sample_size = 30000
total_size = len(df)
test_size_ratio = sample_size / total_size

# Use train_test_split to get stratified indices
_, sampled_indices = train_test_split(df.index, test_size=test_size_ratio, random_state=42, stratify=df['target'])

# Select the rows from the original DataFrame using the sampled indices
df_sample = df.loc[sampled_indices]
print(f"Sampled dataset shape: {df_sample.shape}\n")
```

```python
# -----------------------------------------------
# 3. Data Preprocessing
# -----------------------------------------------
print("Step 3: Data Preprocessing...")

# --- Dataset Info ---
print("--- Dataset Info (on sampled data) ---")
df_sample.info()

# --- Checking for Missing Values ---
print("\n--- Missing Values Check ---")
missing_values = df_sample.isnull().sum().sum()
print(f"Total missing values in the dataset: {missing_values}")

# --- Dataset Description (Statistics) ---
print("\n--- Dataset Description (Top 5 rows) ---")
print(df_sample.head())

# --- Feature Scaling and Splitting ---
# Separate features (X) and target (y)
X = df_sample.drop('target', axis=1)
y = df_sample['target']

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Scale the features. SVMs are highly sensitive to feature scales.
print("\nScaling features using StandardScaler...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Preprocessing complete.\n")
```

```python
# -----------------------------------------------
# 4. Directly Go for the Result (Checking Accuracy)
# -----------------------------------------------
print("Step 4: Training SVM model and checking accuracy...")

# Initialize the Linear Support Vector Classifier
# We use LinearSVC as it's much faster than SVC(kernel='linear') or SVC(kernel='rbf')
# for large datasets. dual=False is recommended when n_samples > n_features.
model = LinearSVC(dual=False, random_state=42, max_iter=5000)
```

```python
# Train the model
print("Training the LinearSVC model... This may take a moment.")
model.fit(X_train_scaled, y_train)
print("Model training complete!")

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\n--- Model Result ---")
print(f"Accuracy of the SVM model: {accuracy * 100:.2f}%")
print("--------------------\n")
```

```python
# -----------------------------------------------
# 5. Represent This Result in Some Graphical Manner
# -----------------------------------------------
print("Step 5: Plotting graphical results (Confusion Matrix)...")

# Get the list of class names (which are numbers 1-7)
class_names = [str(i) for i in sorted(y.unique())]

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

disp.plot(ax=ax, cmap=plt.cm.Blues, xticks_rotation='vertical')
ax.set_title("SVM Confusion Matrix for Forest Cover Type")

print("Displaying plot. Close the plot window to finish the script.")
plt.show()

print("\nScript finished.")
```
